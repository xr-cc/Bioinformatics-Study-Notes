Pattern Counting

```
def PatternCount(Text, Pattern):
    k = len(Pattern)
    cnt = 0
    for i in range(len(Text)-k+1):
        if Text[i:(i+k)]==Pattern:
            cnt +=1
    return cnt
```

Frequent Words Problem

```
def FrequentWords(Text, k):
    cnts = list()
    FrequentPatterns = list()
    for i in range(len(Text)-k+1):
        pat = Text[i:(i+k)]
        cnts.append(PatternCount(Text,pat))
    maxCnt = max(cnts)
    for i in range(len(Text)-k+1):
        if cnts[i]==maxCnt:
            FrequentPatterns.append(Text[i:(i+k)])
    FrequentPatterns = set(FrequentPatterns)
    return sorted(FrequentPatterns)
    
```

Reverse Complement

```
def ReverseComplement(Pattern):
    ATCG_map = {'A':'T','T':'A','C':'G','G':'C'}
    comp = [ATCG_map[i] for i in Pattern]
    r = ""
    for x in comp[::-1]: 
        r += x  
    return r
```

Pattern Matching

```
def PatternMatching(Pattern, Genome):
    k = len(Pattern)
    occ = list()
    for i in range(len(Genome)-k+1):
        if Genome[i:(i+k)]==Pattern:
            occ.append(i)
    return occ
```

Clump Finding

```
def ClumpFinding(genome, k, L, t):

    all_kmers = list()
    for i in range(len(genome)-k+1):
        all_kmers.append(genome[i:(i+k)])
    all_kmers = sorted(set(all_kmers))
    
    is_clump = [False]*len(all_kmers)   
    for i in range(len(genome)-L+1):
        window = genome[i:(i+L)]
        for ik,kmer in enumerate(all_kmers):
            if PatternCount(window,kmer)>=t:
                is_clump[ik] = True
    r = list()
    for ik,kmer in enumerate(all_kmers):
        if is_clump[ik]:
            r.append(kmer)
    return r
```

```
genome = 'CGGACTCGACAGATGTGAAGAACGACAATGTGAAGACTCGACACGACAGAGTGAAGAGAAGAGGAAACATTGTAA'
k = 5 
L = 50 
t = 4
ClumpFinding(genome, k, L, t)
```

Pattern to Number

```
def PatternToNumber(pat):
    allele = ['A','C','G','T']
    pat2num = {k: v for v, k in enumerate(allele)}
    num = 0
    for i,p in enumerate(pat[::-1]):
        num += pat2num[p]*4**i
    return num
PatternToNumber('AGT')
```

Number to Pattern

```
def NumberToPattern(index, k):
    allele = ['A','C','G','T']
    num2pat = {str(v):vv for v, vv in enumerate(allele)}
       
    def numberToBase(n, b):
        digits = [0]*b
        idx = b-1
        while n and idx>=0:
            n,q = divmod(n,4)
            digits[idx] = q
            idx -= 1
        if idx == 0:
            digits[idx] = n
        ret = ""
        for d in digits: 
            ret += str(d)
        return ret

    num = numberToBase(index, k)

    pat = ""
    for l in num:
        pat += num2pat[l]
        
    return pat
```

Computing Frequencies

```
def ComputingFrequencies(text, k):
    
    allele = ['A','C','G','T']
    pat2num = {k: v for v, k in enumerate(allele)}

    FrequencyArray = list()
    for i in range(4**k):
        FrequencyArray.append(0)
    for i in range(len(text)-k+1):
        pat = text[i:(i+k)]
        j = PatternToNumber(pat)
        FrequencyArray[j] += 1
    return FrequencyArray
```
