```
def Skew(genome):
  s = [0]
  numGminusC = 0
  for i in range(len(genome)):
    if genome[i]=='G':
      numGminusC += 1
    elif genome[i]=='C':
      numGminusC -= 1
    s.append(numGminusC)
  return s
```

Minimum Skew Problem

```
def MinimumSkew(Genome):
  skew = [0]
  numGminusC = 0

  minSkew = len(Genome)
  for i in range(len(Genome)):
    if Genome[i]=='G':
      numGminusC += 1
    elif Genome[i]=='C':
      numGminusC -= 1

    if numGminusC<minSkew: # update
      minSkew = numGminusC 

    skew.append(numGminusC)

  minIdx = list()
  for i,s in enumerate(skew):
    if s==minSkew:
      minIdx.append(i)

  return minIdx
```

Hamming Distance Problem

```
def HammingDistance(p, q):
    l = len(p)
    assert(l==len(q))
    dist = 0
    for i in range(l):
      if p[i]!=q[i]:
        dist += 1
    return dist
```

Approximate Pattern Matching Problem

```
def ApproximatePatternMatching(Text, Pattern, d):
    positions = [] # initializing list of positions
    
    k = len(Pattern)
    for i in range(len(Text)-k+1):
      if HammingDistance(Pattern,Text[i:(i+k)])<=d:
        positions.append(i)
        
    return positions
```

Approximate Pattern Count

```
# Input:  Strings Pattern and Text, and an integer d
# Output: The number of times Pattern appears in Text with at most d mismatches
def ApproximatePatternCount(Text, Pattern, d):
    count = 0 # initialize count variable
    k = len(Pattern)
    for i in range(len(Text)-k+1):
      if HammingDistance(Pattern,Text[i:(i+k)])<=d:
        count += 1
    return count
```

d-Neighborhood of a String

```
def ImmediateNeighbors(Pattern):
  neighborhood = list()
  nucleotides = ['A','C','G','T']
  k = len(Pattern)
  for i in range(k):
    symbol = Pattern[i]
    for nuc in nucleotides:
      if nuc!=symbol:       

        if i==0:
          prefix = ""
        elif i==1:
          prefix = Pattern[0]
        else:
          prefix = Pattern[:(i-1)]
        
        if i==(k-1):
          suffix = ""
        else:
          suffix = Pattern[(i+1):k]

        temp = prefix+nuc+suffix
        neighborhood.append(temp)
        
  return sorted(set(neighborhood))
```

```
def Neighbors(Pattern, d):
  nucleotides = ['A','C','G','T']
  if d==0:
    return [Pattern]
  if len(Pattern)==1:
    return nucleotides

  neighborhood = list()
  SuffixNeighbors = Neighbors(Pattern[1:], d)
  for s in SuffixNeighbors:
    if HammingDistance(Pattern[1:], s) < d:
      for nuc in nucleotides:
        neighborhood.append(nuc+s)
    else:
      neighborhood.append(Pattern[0]+s)

  return sorted(set(neighborhood))

```

Frequent Words with Mismatches Problem

```
# first generates all neighbors (with up to d mismatches) for all k-mers in Text 
# and combines them all into an array NeighborhoodArray

def FrequentWordsWithMismatches(Text, k, d):
  FrequentPatterns = list()

  Neighborhoods = list()
  for i in range(len(Text)-k+1):
    for nb in Neighbors(Text[i:(i+k)], d):
      Neighborhoods.append(nb)
  Neighborhoods = sorted(Neighborhoods)
  NeighborhoodArray = list(Neighborhoods)

  Index = list()
  Count = list()
  for i in range(len(Neighborhoods)):
    Pattern = NeighborhoodArray[i]
    Index.append(PatternToNumber(Pattern))
    Count.append(1)
  SortedIndex = sorted(Index)

  for i in range(len(Neighborhoods)-1):
    if SortedIndex[i]==SortedIndex[i+1]:
      Count[i+1] = Count[i]+1
  maxCount = max(Count)

  for i in range(len(Neighborhoods)):
    if Count[i]==maxCount:
      Pattern = NumberToPattern(SortedIndex[i], k)
      FrequentPatterns.append(Pattern)
  
  FrequentPatterns = sorted(set(FrequentPatterns))
  return FrequentPatterns

def PatternToNumber(pat):
    allele = ['A','C','G','T']
    pat2num = {k: v for v, k in enumerate(allele)}
    num = 0
    for i,p in enumerate(pat[::-1]):
        num += pat2num[p]*4**i
    return num

def NumberToPattern(index, k):
    allele = ['A','C','G','T']
    num2pat = {str(v):vv for v, vv in enumerate(allele)}
       
    def numberToBase(n, b):
        digits = [0]*b
        idx = b-1
        while n and idx>=0:
            n,q = divmod(n,4)
            digits[idx] = q
            idx -= 1
        if idx == 0:
            digits[idx] = n
        ret = ""
        for d in digits: 
            ret += str(d)
        return ret

    num = numberToBase(index, k)

    pat = ""
    for l in num:
        pat += num2pat[l]
        
    return pat
```

Frequent Words with Mismatches and Reverse Complements Problem

```
# first generates all neighbors (with up to d mismatches) for all k-mers in Text 
# and combines them all into an array NeighborhoodArray

def FrequentWordsWithMismatchesAndReverseComplements(Text, k, d):
  FrequentPatterns = list()

  Neighborhoods = list()
  for i in range(len(Text)-k+1):
    for nb in Neighbors(Text[i:(i+k)], d):
      Neighborhoods.append(nb)
      Neighborhoods.append(ReverseComplement(nb))
  Neighborhoods = sorted(Neighborhoods)
  NeighborhoodArray = list(Neighborhoods)

  Index = list()
  Count = list()
  for i in range(len(Neighborhoods)):
    Pattern = NeighborhoodArray[i]
    Index.append(PatternToNumber(Pattern))
    Count.append(1)
  SortedIndex = sorted(Index)

  for i in range(len(Neighborhoods)-1):
    if SortedIndex[i]==SortedIndex[i+1]:
      Count[i+1] = Count[i]+1
  maxCount = max(Count)

  for i in range(len(Neighborhoods)):
    if Count[i]==maxCount:
      Pattern = NumberToPattern(SortedIndex[i], k)
      FrequentPatterns.append(Pattern)
  
  FrequentPatterns = sorted(set(FrequentPatterns))
  return FrequentPatterns

def ReverseComplement(Pattern):
    ATCG_map = {'A':'T','T':'A','C':'G','G':'C'}
    comp = [ATCG_map[i] for i in Pattern]
    r = ""
    for x in comp[::-1]: 
        r += x  
    return r
```

```
text = 'ACGTTGCATGTCGCATGATGCATGAGAGCT'
k = 4 
d = 1

FrequentWordsWithMismatchesAndReverseComplements(text, k, d)
```
